Table Customer

```python
Faker.seed(0)
np.random.seed(0)

id = []
f_name = []
l_name = []
gen = ['Male', 'Female', 'Non-binary']
dob = []
occupation = []
# branch_id = []

# Shuffle the branch data based on state_data
shuffled_branch_data = [(k, v) for k, v in state_data.items()]
random.shuffle(shuffled_branch_data)

# This loop iterates through each customer and assigns them to a branch based on the distribution of branches in each state.
# It ensures that customers are realistically distributed across branches by cycling through the states and randomly selecting branch IDs within each state's available branch count.
for _ in range(30000):
    id.append(_)
    f_name.append(fake.first_name())
    l_name.append(fake.last_name())
    dob.append(fake.date_of_birth(minimum_age=18, maximum_age=70))
    gen.append(np.random.choice(gen))
    selected_state, branch_count = shuffled_branch_data[_ % len(shuffled_branch_data)]
    branch_id.append(fake.random_int(min=1, max=branch_count))
    occupation.append(fake.job())
    

df_cust = pd.DataFrame(zip(id, f_name, l_name, gen, dob, branch_id, occupation), columns=["customer_id", "first_name", "last_name", "gender", "date_of_birth", "branch_id", "occupation"])
df_cust.head()
```

Table Card

```python
Faker.seed(0)
np.random.seed(0)

probabilities = [0.7, 0.3]

card_id = []
expiration_dates = []
c_number = []
is_blocked = ["Blocked", "Unblocked"]

for _ in range(30000):
    card_id.append(_)
    expiration_date = fake.date_between(start_date="today", end_date="+10y").strftime("%Y-%m-%d")
    expiration_dates.append(expiration_date)

    card_number = fake.credit_card_number(card_type=None) 
    while len(card_number) != 16:
        card_number = fake.credit_card_number(card_type=None)  # Regenerate the number if it's not 16 digits long
    c_number.append(card_number)

    status = np.random.choice(["Unblocked", "Blocked"], p=probabilities)
    is_blocked.append(status)

df_card = pd.DataFrame(zip(card_id, expiration_dates, c_number, is_blocked), columns=["card_id", "expiration_date", "card_number", "is_blocked"])
df_card.head()
```

Table Account

```python
import pandas as pd
import numpy as np
import random
import datetime
from faker import Faker
from tqdm import tqdm

# Set random seed for reproducibility
Faker.seed(0)
np.random.seed(0)

# Generate account data
account_id = []
balance = []
customer_id = []
card_id = []
# acc_start_date = []

# Shuffle the customer_id list to get random order from customer table
shuffled_customer_ids = id.copy()
random.shuffle(shuffled_customer_ids)

shuffled_card_ids = df_card['card_id'].tolist()
random.shuffle(shuffled_card_ids)


# Generate random dates earlier than the start_date in df_loan
# min_date = min(start_date) - datetime.timedelta(days=365)  # 1 year earlier
# sum=0
for _ in tqdm(range(30000)):
    account_id.append(_)
    beta_value = np.random.beta(5, 2)
    inverted_beta_value = 1 - beta_value  # Invert the distribution to skew towards higher values
    balance_ = 10000 + (inverted_beta_value * (10000 - 0))  # Scale and shift the beta value to the desired range
    balance.append(round(balance_, 2))

    customer_id.append(random.choice(shuffled_customer_ids))
    card_id.append(random.choice(shuffled_card_ids))
    
    # Generate random date earlier than the start_date of loans
    # acc_start_date.append(Faker().date_between(start_date=min_date, end_date=min(start_date)))

# Create DataFrame
df_account = pd.DataFrame(zip(account_id, balance, customer_id, card_id), 
                          columns=["account_id", "bank_balance", "customer_id", "card_id"])
df_account.head()

```

Table Transaction

```python
# Initialize Faker and seeds to ensure reproducibility
Faker.seed(0)
np.random.seed(0)

# Define the lists
transaction_id = []
dates = []
account_ids = []
descriptions = []
transaction_amounts = []

shuffled_account_ids = account_id.copy()
random.shuffle(shuffled_account_ids)

# Generating fake data for each list
for _ in range(30000):  # Assuming we want to generate 5 transactions as in the picture
    transaction_id.append(_)
    dates.append(fake.date_between(start_date='-10y', end_date='today'))
    
    account_ids.append(random.choice(shuffled_account_ids))

    descriptions.append(fake.sentence(nb_words=6))
    beta_value = np.random.beta(5, 2)
    transaction_amount = beta_value * 700  # Scale the beta value to the desired range 0-700
    transaction_amounts.append(round(transaction_amount, 2))  # Round to 2 decimal places

# Create a DataFrame
df_transactions = pd.DataFrame({
    'transaction_id': transaction_id,
    'date': dates,
    'account_id': account_ids,
    'description': descriptions,
    'transaction_amount': transaction_amounts
})

df_transactions.head()
```

Table Loan_Type

```python
Faker.seed(0)
np.random.seed(0)

loan_type_id = []
type = ['Personal', 'Mortgage', 'Auto', 'Student', 'Home Equity', 'Small Business']
description = []
base_amount = []
base_interest_rate = []

for _ in range(30000):
    loan_type_id.append(_)

    loan_type = random.choice(type)
    type.append(loan_type)

    descriptions = fake.sentence(nb_words=12)  # Generates a sentence with 12 words
    description.append(descriptions)

    amount = fake.pydecimal(left_digits=5, right_digits=2, positive=True, min_value=1000, max_value=50000)
    base_amount.append(amount)

    interest_rate = round(np.random.uniform(1.5, 20.0), 2)
    base_interest_rate.append(interest_rate)
    

df_loan_type = pd.DataFrame(zip(loan_type_id, type, description, base_amount, base_interest_rate), columns=["loan_type_id", "loan_type", "loan_type_description", "base_amount", "base_interest_rate"])
df_loan_type.head()
```

Table Loan

```python
Faker.seed(0)
np.random.seed(0)

shuffled_account_ids = df_account['account_id'].tolist()
random.shuffle(shuffled_account_ids)

shuffled_loan_type_ids = df_loan_type['loan_type_id'].tolist()
random.shuffle(shuffled_loan_type_ids)

loan_id = []
start_date = []
due_date = []
amount_paid = []
account_id = []
loan_type_id = []

for _ in range(30000):
   loan_id.append(_)

   start = fake.date_between(start_date='-10y', end_date='today')
   start_date.append(start)
   
   due = fake.date_between(start_date=start, end_date='+5y')
   due_date.append(due)
    
   # paid = round(random.uniform(0, float(base_amount[_]) * 0.9), 2)
   # amount_paid.append(paid)
   paid = round(random.uniform(0, float(base_amount[_]) * 0.9), 2)
   amount_paid.append(paid)

   account_id.append(random.choice(shuffled_account_ids))

   loan_type_id.append(random.choice(shuffled_loan_type_ids))

df_loan = pd.DataFrame(zip(loan_id, account_id, loan_type_id, due_date, start_date, amount_paid), columns=["loan_id", "account_id", "loan_type_id", "due_date", "start_date", "amount_paid"])
df_loan.head()
```
